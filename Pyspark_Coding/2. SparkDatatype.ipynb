```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('Word Count').getOrCreate()
```

```python
data = [
	{
		"color": "red",
		"value": "#f00"
	},
	{
		"color": "green",
		"value": "#0f0"
	},
	{
		"color": "blue",
		"value": "#00f"
	},
	{
		"color": "cyan",
		"value": "#0ff"
	},
	{
		"color": "magenta",
		"value": ""
	},
	{
		"color": "yellow",
		"value": None
	},
	{
		"color": "black",
		"value": "#000"
	}
]
```

```python
from pyspark.sql import Row
```

```python
data_row = [Row(**i) for i in data]
```

```python
df = spark.createDataFrame(data_row)
df.show()
```

#None is converted to null 
"" will be empty
